### Q2: Continue "Q2" of the previous weeks 3 and 4

1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

3. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a yet another different again non normal distribution and your prior for $\tau$ is a yet another different again non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import numpy as np
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt

# Simulated data
np.random.seed(42)
n = 30
true_theta = 5
true_tau = 1  # Precision
data = np.random.normal(true_theta, 1 / np.sqrt(true_tau), size=n)

# Number of samples for MCMC
n_iter = 2000
burn_in = 500

# 1: Normal prior for theta, Gamma prior for tau
with pm.Model() as model_1:
    theta = pm.Normal("theta", mu=0, sigma=2)  # Normal prior
    tau = pm.Gamma("tau", alpha=2, beta=1)     # Gamma prior
    
    likelihood = pm.Normal("obs", mu=theta, tau=tau, observed=data)
    
    trace_1 = pm.sample(n_iter, return_inferencedata=True, tune=burn_in, chains=2, target_accept=0.9)

# 2: Laplace prior for theta, Exponential prior for tau
with pm.Model() as model_2:
    theta = pm.Laplace("theta", mu=0, b=2)  # Laplace prior
    tau = pm.Exponential("tau", lam=1)      # Exponential prior
    
    likelihood = pm.Normal("obs", mu=theta, tau=tau, observed=data)
    
    trace_2 = pm.sample(n_iter, return_inferencedata=True, tune=burn_in, chains=2, target_accept=0.9)

# 3: Student(t) prior for theta, Half-Normal prior for tau
# We use Hald-Normal since tau should be greater than 0
with pm.Model() as model_3:
    theta = pm.StudentT("theta", nu=3, mu=0, sigma=2)  # Student-T prior
    tau = pm.HalfNormal("tau", sigma=1)               # Half-Normal prior
    
    likelihood = pm.Normal("obs", mu=theta, tau=tau, observed=data)
    
    trace_3 = pm.sample(n_iter, return_inferencedata=True, tune=burn_in, chains=2, target_accept=0.9)

# Plot trace for all cases
az.plot_trace(trace_1)
plt.suptitle("Case 1: Normal Prior for Î¸, Gamma Prior for Ï„", fontsize=14)
plt.show()

az.plot_trace(trace_2)
plt.suptitle("Case 2: Laplace Prior for Î¸, Exponential Prior for Ï„", fontsize=14)
plt.show()

az.plot_trace(trace_3)
plt.suptitle("Case 3: StudentT Prior for Î¸, Half-Normal Prior for Ï„", fontsize=14)
plt.show()

# Summary statistics
summary_1 = az.summary(trace_1)
summary_2 = az.summary(trace_2)
summary_3 = az.summary(trace_3)

print("Summary Statistics for Case 1:")
print(summary_1)

print("\nSummary Statistics for Case 2:")
print(summary_2)

print("\nSummary Statistics for Case 3:")
print(summary_3)
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

```python
import pymc
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)    
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" 
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

To start the explanation, augments are the first elements to look at. f: the density function we are sampling from; x: the current sample; y: randomly chosen height base on the former sample. \\
1. Find a, b in [0, 1] such that in [a, b], f(x) > y. \\
2. Get a random sample from [a, b] which is x_. \\
3. if x_ is larger than y, return x_ with count of tries: 1
4. if x_ is smaller thany, shrink the interval based on x, and repeat steps in 2 and 3 with new count of tries

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')
    
plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```

We shall go through the code line by line. 
1. Initialize the target distribution (Beta distribution). Initialize m: number of iteration, x: sample storage, the initial two values of x are 0.25. 
2. by plot_trace, we visualize the first 10 iterations. According to our plot, all first 10 iterations fall into the beat distribution. 
3. (Iterate through Markov chain steps)from 1 to m(m = 1000 in our code), generate f(x) base on current x in range[0, f(x)], do slice-sampling by function slice_f_at_y to find next sample x with it's count of tries.

When we replace MH by slice sampling, it doesn't require tuning of the proposal distribution since we are using slicing instead of Accept Probability to decide.
